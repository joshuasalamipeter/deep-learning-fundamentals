### **1. How GANs and VAEs Work**

In this section, we briefly revisit the two foundational families of generative models we will encounter throughout Project 12: **GANs** and **VAEs**.
Our goal is *not* to derive equations — we want to understand the **intuitive story** behind how these systems create synthetic images that look uncannily real.

Generative models differ from discriminative ones because instead of deciding *what* something is (cat vs. dog), they learn to **produce new samples** that resemble a training distribution.

**1.1 Generative Adversarial Networks (GANs): A Two-Player Game**

GANs are built from two competing neural networks:

* **Generator (G):** Tries to create fake images that look real.
* **Discriminator (D):** Tries to detect whether an image is fake or real.

They learn through a *game of deception*.
If the discriminator becomes good at spotting fakes, the generator must improve.
If the generator becomes good at fooling the discriminator, the discriminator must adapt.

Over many rounds, the generator learns to produce images so realistic that **humans often can’t detect the difference**. [11]

<img src="NB02_assets/06_from_paper_aStyleBasedGenArchitectureforGAN.png" width="550">

Think of GAN training like two artists competing:

* One artist (the generator) draws counterfeit paintings.
* The other (the discriminator) is an art critic attempting to identify forgeries.
* Every time the critic catches the counterfeits, the artist improves.
* Every time the artist fools the critic, the critic sharpens their judgment.

This adversarial loop pushes both networks to improve until the generator’s outputs resemble realistic photos.

We can visualize this with a classic conceptual architecture [12]:

<img src="NB02_assets/13_GANarchitecture_fromPaper_Generative_Adversarial_Networks_for_Synthetic_Data_Generation_A_Comparative_Study.png" width="550">

**1.2 Variational Autoencoders (VAEs): Learning Distributions, Not Deception**

While GANs rely on adversarial dynamics, **VAEs** work differently.
They learn to compress images into a *latent space* and then reconstruct them.

A VAE has two main components:

* **Encoder:** Compresses an image into a low-dimensional “latent code”
* **Decoder:** Reconstructs an image from that latent code

But unlike a standard autoencoder, VAEs learn **distributions** (mean & variance), allowing them to sample new points from the latent space and generate new images [13].

<img src="NB02_assets/14_VAEarchitecture_from_learnopencv_com_variational-autoencoder-in-tensorflow.png" width="550">

A VAE tries to answer the question:

> “What does the *space* of possible faces (or objects) look like?”

Instead of memorizing images, VAEs learn a smooth, continuous latent space where:

* Nearby points correspond to visually similar images
* Interpolation between points creates *plausible but new* images

This ability to **sample** from the latent space is what makes VAEs generative.

GANs are often sharper and more photorealistic, but VAEs remain important because they are:

* Stable to train
* Mathematically principled
* Useful for downstream tasks like disentanglement and representation learning

**1.3 Putting It Together: Two Paths to Synthesis**

| Model Type | How It Generates                     | Strengths                              | Weaknesses                          |
| ---------- | ------------------------------------ | -------------------------------------- | ----------------------------------- |
| **GANs**   | Fooling a discriminator              | Sharp, realistic images                | Training instability, mode collapse |
| **VAEs**   | Sampling from a learned distribution | Stable training, good latent structure | Slightly blurrier outputs           |

Both approaches underpin modern models such as **StyleGAN**, **VAE-GAN hybrids**, and many diffusion architectures.

### **3. Deepfakes: How They Are Made and Why They Matter**

Deepfakes represent one of the most powerful — and potentially harmful — applications of modern generative AI. By combining face-swapping, lip-syncing, voice-cloning, and GAN-based generation, deepfakes can produce audio–visual content that is almost indistinguishable from real footage. 

In this section, we explore *how* deepfakes are made, *why* they work so well, and *what kinds of risks* they pose to individuals, institutions, and society.

**3.1 How Deepfakes Are Made (High-Level Workflow)**

A typical deepfake creation pipeline involves:

1. **Collecting training footage**  
   A model is trained on many frames of a target person's face to capture expressions, angles, and lighting.

2. **Encoding facial features**  
   The model learns the target’s identity in latent space (e.g., eye shape, jawline, wrinkles, expression range).

3. **Generating synthetic frames**  
   A generator produces new frames where another actor’s face is replaced with the target face.

4. **Post-processing**  
   Color matching, smoothing, voice cloning, lip-sync alignment.

> **Why deepfakes look convincing:**  
> Human brains rely heavily on facial cues. When GANs learn these cues — eye blinks, muscle movements, micro-expressions — the output appears “psychologically real,” even if it is technically synthetic.

**3.2 Example 1 — Deepfake Financial Scam (Elon Musk Fraud Video)** [14]

<img src="NB02_assets/09_elonMask_deepFake_FinancialFraudulantPlatform_RMITeduAU_elon-musk-used-in-fake-ai-videos-to-promote-financial-scam.png" width="450">

Online scammers used a deepfake of Elon Musk to promote a fraudulent cryptocurrency and stock-trading platform. In the video, “Musk” appears to endorse the investment scheme and invites viewers to deposit money — none of which he actually said.

**Why this is dangerous:**

- **Authority misuse:** People trust familiar public figures.
- **Visual authenticity:** The video uses a real background and realistic lighting.
- **Scalable fraud:** AI makes it easy to mass-produce many scam videos.
- **Victim impact:** Viewers lost money believing the message came from Musk.

> **Key lesson:**  
> Deepfakes enable *identity theft at scale*, where trust in well-known figures becomes a vulnerability.

**3.3 Example 2 — Real vs Deepfake (Obama Demonstration)** [15]

<img src="NB02_assets/10_realVSdeepfakeObama_engineering_nyu_edu_news_deepfake-videos-are-getting-better-and-pentagon-worried.png" width="450">

In this well-known research demonstration, Barack Obama’s face and voice were synthetically manipulated to produce entirely fabricated speech.

The image above shows:

- **Left:** Real footage  
- **Right:** Deepfake output  
- **Red dots:** Facial landmark tracking  
- **Blue 3D box:** Head pose estimation  

The deepfake closely matches Obama’s expressions, blinking, and head movements, creating a persuasive illusion.

**Ethical risks:**

- **Political misinformation:** Fake speeches, statements, or endorsements.
- **Election interference:** Altered videos can spread faster than fact-checks.
- **Erosion of trust:** If any video can be fake, all videos become suspicious.

> **Societal implication:**  
> When audiovisual evidence loses credibility, democratic processes and journalism are put at risk.

**3.4 Example 3 — Simple Digital Manipulation (Benign but Important)** [16]

<img src="NB02_assets/15_manipulatedImage_researchgate_A_Dataset_of_Photos_and_Videos_for_Digital_Forensics_Analysis_Using_Machine_Learning_Processing.png" width="450">

Even simple manipulations — like moving an electrical socket from one part of a wall to another — illustrate an important point:

- **Not all harmful media is obviously dramatic.**
- **Small edits can create misleading evidence**, affecting legal cases, insurance claims, workplace disputes, or digital forensics.

Deepfakes are simply the *next step* — extremely sophisticated forms of the same manipulation principle.

**3.5 Why Deepfakes Matter (Broader Ethical + Societal Impact)**

Deepfakes introduce several categories of harm:

**1. Individual Harm**
- Reputation damage  
- Non-consensual deepfake pornography  
- Identity theft  
- Psychological distress  

**2. Social Harm**
- Harassment  
- Cyberbullying  
- Impersonation scams  

**3. Institutional Harm**
- Fake political speeches  
- Military or diplomatic misinformation  
- “Fake evidence” inserted into investigations  

**4. Epistemic Harm (“Truth Decay”)**
- If fake media becomes indistinguishable from real,  
  **the default reaction becomes distrust** — a major threat to democratic societies.

### **4. Creativity vs. Manipulation: Ethical Tensions**

In this section we explore the *double-edged nature* of generative models. The same tools that enable artistic creativity, accessibility, and innovation can also blur the boundaries between reality and fabrication. We discuss both sides carefully, using visual examples.

**4.1 Generative Models as Creative Tools**

Generative models such as GANs and diffusion models have opened entirely new possibilities for artists, designers, filmmakers, and hobbyists. These systems can create original compositions, remix artistic styles, and even democratize access to visual creativity for people without formal artistic training.

**AI Artwork Example — “Imagined Mona Lisa”** [13]

This image is *not* a photograph or a scan of an existing artwork — it was synthesized by a diffusion model.  
It demonstrates how AI can recreate the **style** of classical painters, producing results that feel familiar yet newly imagined.

Such works raise important questions:

- *Is AI-generated art “original”?*  
- *Does stylistic imitation count as creativity or copying?*  

These questions are now being debated in art schools, copyright law, and online creator communities.

**4.2 When Synthetic Media Wins Human Contests**

A well-known example occurred when an AI-generated image won first place at the **Colorado State Art Fair**. 

<img src="NB02_assets/12_uk_pcmag_com_news_142396_man-boasts-about-winning-art-competition-using-ai-created-image.png" width="450">

This example captures a broader societal tension:  
**Should AI works compete with human-made art?**

**4.3 Beneficial Manipulations: Enhancement, Restoration & Accessibility**

Generative models also support positive uses such as historical restoration, accessibility enhancement, and film production.

**(a) Film De-Aging / VFX** [18]

<img src="NB02_assets/18_deAging_moople_in_blog_what-is-de-ageing-and-how-learning-vfx-for-films-can-help-in-de-ageing.png" width="350">

De-aging technologies allow filmmakers to depict characters across timelines, revive performances, or maintain continuity in long-running franchises. These VFX techniques rely on deep generative models to predict younger facial structures while preserving identity.

Benefits include:
- creative storytelling  
- reduced need for prosthetics  
- accessibility for indie creators who cannot afford advanced makeup pipelines  

But risks remain when such techniques are misused (e.g., to fabricate evidence).

**(b) Historical Restoration and Colorization** [19]

<img src="NB02_assets/19_restoration_tugraz_at_news_article_ki-algorithmus-bringt-farbe-in-schwarz-weiss-filme.png" width="500">

- bring damaged or faded photographs back to life,  
- restore old film frames,

Here AI functions as a **tool for preservation**, not deception.

**4.4 The Ethical Tension: Creativity vs. Authenticity**

The same generative power that helps artists or historians can also enable **misinformation, impersonation, and style theft**.

Key tensions include:

- **Authenticity:** When synthetic images become indistinguishable from real ones, how do we verify truth?

- **Style Ownership:** Artists argue their styles are being replicated without consent, raising copyright and attribution questions.

- **Cultural Impacts:** Generative tools may homogenize art or reproduce dominant aesthetic biases.

- **Emotional & Social Risks:** Synthetic media that mimics deceased family members, public figures, or private individuals can impact trust, grief, and identity.

Generative AI does not only create — it *reshapes expectations* about what is real, what is possible, and what is ethically acceptable.

### **5. Case Studies of Ethical Concerns**

In this section, we explore real-world scenarios where synthetic media has directly influenced public perception, political discourse, and our shared understanding of history. These cases help us see how powerful generative technologies become when combined with high-stakes contexts such as elections, wars, and cultural memory. They also show why governance, transparency, and media literacy matter more than ever.

**5.1 Case Study A — Political Manipulation Through Deepfakes**

Political deepfakes are increasingly used to spread misinformation, create confusion, or manipulate public emotions. In moments of geopolitical tension, even a single realistic fake can shape public opinion or escalate conflict.

**Example: AI-generated video falsely depicting U.S. President Joe Biden.** [20]

<img src="NB02_assets/16_fakePoliticalcontent_newsweek_com_deepfake-biden-video-hypes-false-narrative-us-sending-troops-ukraine-1785147.png" width="450">

In this example, an AI-generated video circulated online claiming that President Biden announced the deployment of U.S. troops to Ukraine.  
Nothing in the video was real — but the production quality was high enough that thousands of viewers believed it.

**Why this matters:**

- Such videos exploit the *trust* we place in familiar political figures.  
- They can spread rapidly before fact-checkers intervene.  
- In conflict zones or polarized societies, they can directly influence behaviour (panic, outrage, retaliation).  
- Political deepfakes challenge the integrity of democratic processes and informed citizenship.

**5.2 Case Study B — Fabricating Historical Events**

Generative models are now capable of producing images that look convincingly “old,” complete with film grain, sepia tones, and vintage textures. This capability makes it easy to create fake historical photos that never existed — altering our collective memory.

**Example: AI-generated “historical” image of Henry Ford that never existed.** [21]

<img src="NB02_assets/17_fakeHistoricalPhoto_aap_com_au_factcheck_ai-used-to-generate-fake-history-photos.png" width="450">

At first glance, the photo might seem like a genuine archival image.  
In reality, the person in the image never posed for such a picture — and the vehicle depicted did not match any real historical record.

**Why this matters:**

- Historical misinformation can reshape how societies understand their past.  
- Fake archival images can be weaponised to promote ideologies, rewrite narratives, or support conspiracy theories.  
- The line between authentic documentation and synthetic fabrication becomes dangerously blurred.

**5.3 Key Lessons from These Cases**

Across both examples, the risks are not technical — they are *societal*:

- **Speed > Verification:** Synthetic content spreads faster than corrections.  
- **Realism → Trust:** People tend to trust visual evidence more than text, making visual deepfakes especially persuasive.  
- **Attribution is difficult:** Without clear watermarking or provenance tracking, it’s nearly impossible for an average viewer to distinguish real from synthetic.  
- **Platform amplification:** Social media algorithms may boost sensational fake content more than verified information.  

These cases remind us that generative models operate not just as creative tools but as **actors within information ecologies**.  
Understanding their ethical implications requires us to consider both *technological capability* and *societal vulnerability*.

### **6. Final Reflection: Consent, Authenticity, and Intellectual Property**

This final reflective section invites us to pause and consider the deeper ethical questions raised by generative models and synthetic media. While the technical details of GANs, VAEs, and diffusion models help us understand *how* synthetic content is created, the moral and societal questions help us understand *what it means* for humans, institutions, and culture.
